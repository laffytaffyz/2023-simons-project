```python
import numpy as np
import csv
import os

n = 501  # grid size
n_int = 50  # subgrid size
dmch = 5. / (n - 1)
dq = 23. / (n - 1)
mch = np.arange(n) * dmch  # mch range
q = 12 - np.arange(n) * dq  # q range

csv_arr = []  # initialize csv array 

# mch_0 loop
for i1 in range(n_int + 1):
    mch_0 = 0.5 + 4. / n_int * i1

    # q_0 loop
    for i2 in range(n_int):
        q_0 = 4 - 7 / n_int * i2
        print(i1, i2)

        # q error loop
        for j in range(13):
            e_q = 1 + 0.25 * j
            e_mch = 0.01

            # definition of dp
            dp = np.exp(-0.5 * ((mch[:, np.newaxis] - mch_0) / e_mch) ** 2 - 0.5 * ((q - q_0) / e_q) ** 2) / (2 * np.pi * e_mch * e_q)

            # m1 and m2 axis calculation
            m1 = np.zeros((n, n))
            m2 = np.zeros((n, n))
            realq = np.exp(q) + 1
            for k in range(n):
                m1[k, :] = mch[k] * (1 + realq)**0.2 * realq**0.4
                m2[k, :] = m1[k, :] / realq

            # merger probability calculation
            p_bns = np.sum(dp[m1 < 3] * dmch * dq)
            p_bhns = np.sum(dp[(m1 > 5) & (m2 < 3)] * dmch * dq)
            p_bbh = np.sum(dp[m2 > 5] * dmch * dq)
            p_gap = np.sum(dp[((m1 > 3) & (m1 < 5)) | ((m2 > 3) & (m2 < 5))] * dmch * dq)
            p_ns = np.sum(dp[m2 < 2.25] * dmch * dq)

            # append data to csv_arr
            csv_arr.append([mch_0, q_0, e_q, p_bns, p_bhns, p_bbh, p_gap, p_ns])

            print(p_bns, p_bhns, p_bbh, p_gap, p_ns)

# write csv file
header = ["p_bns", "p_bhns", "p_bbh", "p_gap", "p_ns"]
with open("second_update_mprobs.csv", mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(header)
    writer.writerows(csv_arr)
```
